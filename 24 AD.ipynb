```python
import pandas as pd
text=['Sarah lives in a hut in the village.','She has an apple tree in her backyard.','The apples are red in colour.']
df=pd.DataFrame(text,columns=['Sentence'])
df
```

```python
import spacy 
nlp=spacy.load("en_core_web_sm")
token=[]
pos=[]
for sent in nlp.pipe(df['Sentence']):
    if sent.has_annotation('DEP'):
        token.append([word.text for word in sent])
        pos.append([word.pos_ for word in sent])
```

```python
token
```

```python
df['token']=token
df['pos']=pos
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

# Define the variable 'text' (you need to provide some input text here)
text = "This is an example sentence for testing purposes. Natural Language Processing is fun!"

# Ensure necessary resources are downloaded
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')

stop_words = set(stopwords.words('english'))
tokens = sent_tokenize(text)

for i in tokens:
    words = nltk.word_tokenize(i)
    words = [w for w in words if not w.lower() in stop_words]  # Fixed to handle case-insensitivity
    tags = nltk.pos_tag(words)  # 'tags' will be updated in each loop iteration

tags  # Output tags (this will contain only the last sentence's tags)
```
